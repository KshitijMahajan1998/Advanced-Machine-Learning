# Question 2 : t-SNE (7 pts)

Read this [article](https://erdem.pl/2020/04/t-sne-clearly-explained) on t-SNE, a dimensionality reduction technique for visualization and explain it in your own words in one or two paragraphs.

## Solution:

T-SNE is one of the techniques for dimensionality reduction that helps us in understanding the associations/correlations between different features.
Some of the key features of T-SNE is,

1) It is <b>Iterative</b>

2) It is <b>non deterministic</b>

3) It is used for data which is <b>not linearly seperable</b>

4) It is specially used when we have <b>high dimensional data</b> that is to be projected into a low dimensional space

Elaborating more on T-SNE,

When dealing with data in an N dimensional space, we have data that is clustered together and thus having many locally neighbouring groups. Because of being in an N dimensional space, that data **cannot be seperated by a straight line.** This complexity does not allow the reduced matrix obtained from T-SNE to be projected/applied on another dataset. Sometimes when we run T-SNE on the same training dataset we might get different reduced matrices. This makes it <b>less useful for training datasets in ML.</b> This is not applicable for PCA but then PCA can't handle extreme high dimensions too. Thus T-SNE is useful for CNN which has high complexity and multiple layers.

Now looking at how T-SNE works,

* In layman's terms, we will be **finding out similarity of each point** from all other points and repeat this process for all points. This gives an **array of similarity score of one point with each other point.** Now combining all these arrays for each point we get a square matrix of similarity scores.

* To calculate the similarity of each point, we generate a **probability distribution that represent similarity values with all neighbours.** It is a normal curve following a **gaussian distribution.** Each point on the x axis is a neighbour to the desired point with the **unscaled similarity score on the curve.** Close points have high similarity values and far off values have low similarity values.

* But there is a slight issue in this process, there comes situations, where, the data points might be **more dense** and hence the normal curve which we obtain might be having **more standard deviation** for some point and less standard deviation for some points. That makes comparison difficult because the similarity value for two points with respect to each other might be different when computed by each of them in opposite directions. This is because they are **following different distributions.** 

* Hence, we **scale the similarity values on a scale of 0 to 1.** We divide the score by all similarity scores for that point which gives us a scaled score. and in between two points when computing score with each other in opposite directions, **we average it over N dimensions**, thus giving a generic scaled value.

* This way we generate our similarity matrix for all points

* This bring us to another important parameter, **PERPLEXITY**, which is nothing but the **expected density of the points** in other words what is the **expected number of neighbours we wish to achieve in a cluster.** The **greater the preplexity, higher the variance** in our scaled similarity distribution. It's value **ranges in between 5 and 50** and is chosen before running this algo. It is positively correlated with the Mean of the distribution.

* Next task is to **generate a low dimensional projections.** For that we distribute our point in a low dimensional space, maybe 1 or 2. We then **calculate unscaled similarity distribution for these points randomly spread.** For this task here instead of using gaussian distribution we use **student T distribution in single degree because it has a very long tail**, as the name says.

* **A gradient** is calculated for each point which describes how far or close it is to the **point of strongest affinity/attraction.** The point moves in the direction with the strongest pull.

* Each time the point moves one at a time in a specific direction and we end up getting a matrix similar to the matrix obtained in high dimension. 

Some optimizations that T-SNE performs are,

1) *Early Compression:* It levies an L2 penalty on the similarity score to avoid fast clustering of point towards their local neighbouring groups.

2) *Early Exaggeration:* By using conditional probabilities, we avoid data points while moving to bump into each other, meaning we avoid the scores to coincide by spreading it and scaling it evenly.


# Question 3: PCA (8 pts)



Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



## Solution:


Principle Component Analysis is an **algorithm used for dimensionality reduction of your feature space** working on the concept of **feature extraction** so you have less relationships to understand. It constructs **new variables out of your old variables by drawing linear combinations of them**. These linear combinations are made in such a way that the new variables are independent and have no correlation in between them. This allows us to use these new variables further in linear regression as well.

Maximum information gets stored in the few starting components(new variables) and this **allows us to eliminate some variables from the back depending on how much information we are interested in,** then we can drop the least important variables while still **retaining the most valuable information of all of the variables!**

So with **n dimensional data we get n principal components,** but as mentioned we can focus our analysis and regression on the first few components without suffering any great loss which we would have had we removed any feature in our old variable set.

What PCA does is, it **finds the best fitted line by maximizing the sum of the squared distances from the projected points to the origin on a coordinate system** where each data point is plotted with respect to its value along each dimension. This line is what we call principal component 1. Similar to this, along each dimension we construct a perpendicular line to the previous principle component, and hence we **obtain many components acting as our new variables.**

Each line has an associated slope to it which is represented by a vector which also consists of the direction of that line. This vector is what we call **eigan vector** and so we have each eigan vector representing each component. **Eigan Values represent the magnitude of each component** thus providing us with the variations of the feature space.
If there are eigenvalues close to zero, they represent components that may be discarded.

The eigenvectors and eigen values are calculated from the covariance matrix of the feature space and so the eigenvectors where eigen values of that particular component are high are selected as those with high eigen values are the ones which cover most of the variance in the feature space. Because **we can easily remove components that represent least variations in between the data,** the **eigan vectors which are representing principal components are the perfect choice to present optimal data.**

Lastly,
The Principal Component values are calculated from by **multiplying the matrix of eigen vectors with the feature space to find out variations in between features.** To reconstruct the original data we need to multiply the inverse of the eigan vectors matrix with Principal Component values or eigan values. Eigan values are nothing but a squared contribution of the scores on a unite level. If there is any **missing information in the values of Principal Components** then we will get a **noisy version of the original data.**

for eg: we use this to get prinicpal components, 
$$ PC = X.V $$
where PC are principal components, X is data, and V are scores from eigan vectors. Therefore to get original noisy data we can do, 

$$X_{noisy} = PC.V^T$$
$$OR$$
$$X_{noisy} = PC.V^ -1$$

## Part E

1) **(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

2) **(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
# Part (i)

np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
sample_std = sample.std()
z_critical = round(st.norm.ppf(.975), 2)#1.96                       # 95% ‚âà 1.96 standard deviations

#Get population standard deviation
pop_stdev = data_dm["SalePrice"].std()

#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print("True mean value: {mean}\n95% confidence interval: {lower_bound} - {upper_bound}".format(mean = sample_mean,
                                                                                            lower_bound = round(confidence_interval[0],5),
                                                                                            upper_bound = round(confidence_interval[1],5)))
````

````python
# Part (ii)
np.random.seed(12)

sample_size = 500

intervals = []
sample_means = []

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    # compute z critical value
    sample_std = sample.std()
    z_critical = round(st.norm.ppf(.975), 2)#1.96                       # 95% ‚âà 1.96 standard deviations

    #Get population standard deviation
    pop_stdev = data_dm["SalePrice"].std()

    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
    
plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for bot,top in intervals],
             fmt='o')

# plt.hlines(xmin=0, xmax=25,
plt.hlines(xmin=0, xmax=100,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````

### Observation:

Confidence interval of a sample is the "allowable deviation" from the population, such that it can be still considered to have been derived from the same population. If the sample mean is too far off,  hypothesis (that the sample belongs to the population) gets rejected.

In the above picture, **the horizontal line shows the population mean while the vertical line represents the sample mean and the 1.96*standard deviation value of the sample.**
